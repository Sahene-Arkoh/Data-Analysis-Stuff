```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168')
```

```python
page.status_code
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
seven_day=soup.find(id='seven-day-forecast')
```

```python
forecast_items=seven_day.find_all(class_='tombstone-container')
```

```python
today=forecast_items[0]
```

```python
print(today.prettify())
```

```python
period=today.find(class_='period-name').get_text()
shortdesc=today.find(class_='short-desc').get_text()
temp=today.find(class_='temp').get_text()
```

```python
print(period)
print(shortdesc)
print(temp)
```

```python
img=today.find('img')
desc=img['title']
print(desc)
```

```python
period_tags=soup.select('.tombstone-container .period-name')
periods=[pt.get_text() for pt in period_tags]
periods
```

```python
short_descs=soup.select('.tombstone-container .short-desc')
s_descs=[d.get_text() for d in short_descs]
s_descs
```

```python
temps=[t.get_text() for t in soup.select('.tombstone-container .temp')]
```

```python
Descs=[im['title'] for im in soup.select('.tombstone-container img')]
```

```python
forecast=pd.DataFrame({'Period':periods, 'Short desc':s_descs, 'Temp':temps, 'Image':img})
```
